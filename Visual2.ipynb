```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import random
import warnings

warnings.filterwarnings('ignore')
data = pd.read_csv("D:/Titanix/DataSet/Iris.csv")
```

```python
data.head()
```

```python
data.drop(['Id'],axis = 1, inplace = True)
```

```python
data.head()
```

```python
data.drop(['Species'],axis=1 , inplace = True)
```

```python
data.head()
```

```python
data['PetalWidthCm'].unique()
```

```python
data.columns = ['SL','SW','PL','PW']
```

```python
data.tail()
```

```python
plt.figure(1)
plt.plot(data.index,data['SL'],label = 'Sepal Length')
plt.plot(data.index,data['SW'],label = 'Sepal Width')
plt.plot(data.index,data['PL'],label = 'Petal Length')
plt.plot(data.index,data['PW'],label = 'Petal Width')
plt.xlabel('Index')
plt.ylabel('Distribution of Val')
plt.title('Iris Flowers')
plt.legend()
plt.show()
```
