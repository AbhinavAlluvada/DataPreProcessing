```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
axes = plt.axes(projection = '3d')
arr = np.random.randint(1,50,[500,])
arr2 = np.random.randint(1,50,[500,])
axes.scatter(arr,arr2,marker='*',color ='red',alpha=0.6)
axes.view_init(azim = 0,elev = 90)
plt.show()
```

```python
axes = plt.axes(projection = '3d')
val = np.random.normal(15,1.5,100)
val1 = np.random.normal(15,1.5,100)
val2 = np.random.normal(15,1.5,(100,100))

X, Y = np.meshgrid(val, val1)

axes.plot_surface(X,Y,val2,cmap='plasma')

plt.show()
```

```python
axes = plt.axes(projection = '3d')
val = np.random.normal(15,1.5,100)
val1 = np.sin(val)
val2 = np.random.normal(15,1.5,(100,100))

X, Y = np.meshgrid(val, val1)

axes.plot_surface(X,Y,val2,cmap='plasma')

plt.show()
```

```python
axes = plt.axes(projection = '3d')

x = np.arange(-5,5,0.1)
y = np.arange(-5,5,0.1)

X,Y = np.meshgrid(x,y)
Z = np.sin(X) * np.cos(Y)

axes.plot_surface(X,Y,Z,cmap='plasma')

plt.show()
```
